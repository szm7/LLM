```python
data = pd.read_csv('/Users/saninzulphi/Downloads/Articles.csv', encoding='latin-1')
```

```python
import re
```

```python
# Text cleaning function
def clean_text(text):
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove non-alphabetic characters
    text = text.lower()  # Convert to lowercase
    return text

# Clean the lyrics
data['clean_article'] = data['Article'].apply(clean_text)
```

```python
from transformers import BertTokenizer, BertModel

# Load the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Load the BERT model
#model = BertModel.from_pretrained('bert-base-uncased')
```

```python
def tokenize_and_encode(text):
    encoding = tokenizer(text, padding='max_length', truncation=True, return_tensors='pt')
    return encoding['input_ids'], encoding['attention_mask']

data['input_ids'], data['attention_masks'] = zip(*data['Article'].apply(tokenize_and_encode))

print(data[['Article', 'input_ids', 'attention_masks']].head())
```

```python
from transformers import AutoTokenizer, AutoModel

# Load the BioBERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-base-cased-v1.1")
model = AutoModel.from_pretrained("dmis-lab/biobert-base-cased-v1.1", output_hidden_states=True)

# Tokenize a sentence
text = "Cancer is a disease that affects cells in the body."
inputs = tokenizer(text, return_tensors="pt")

# Get the contextual embeddings
outputs = model(**inputs)

# Access the hidden states
hidden_states = outputs.hidden_states
embeddings = hidden_states[-1].squeeze(0)

print(embeddings)
print(inputs)
```

```python
from transformers import AutoTokenizer, AutoModel

# Load the ClinicalBERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
model = AutoModel.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")

# Tokenize a sentence
text = "The patient was diagnosed with pneumonia and started on antibiotics."
inputs = tokenizer(text, return_tensors="pt")

# Get the contextual embeddings
outputs = model(**inputs)
embeddings = outputs.last_hidden_state.squeeze(0)

print(embeddings)
print(inputs)
```
